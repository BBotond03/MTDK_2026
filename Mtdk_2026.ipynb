```python
import os
import shutil
from google.colab import drive

#drive.mount('/content/drive', force_remount=True)

# 1. Mount Google Drive
# This prompts whoever is running the cell to log in and grant access to their Drive.
drive.mount('/content/drive')

# 2. Define the paths
# This assumes you made a shortcut in your main Drive named exactly "Mtdk 2026"
shared_folder_shortcut = '/content/drive/MyDrive/Mtdk__2026'

# This is the new folder that will be created in your Drive to hold your personal working copies
my_workspace_folder = '/content/drive/MyDrive/Mtdk_2026_Working_Files'

# Create the destination folder if it doesn't already exist
os.makedirs(my_workspace_folder, exist_ok=True)

# 3. Copy files from the shared folder to the new workspace folder
print(f"Checking for files in your shortcut: {shared_folder_shortcut}...\n")

if os.path.exists(shared_folder_shortcut):
    for item in os.listdir(shared_folder_shortcut):
        source_item = os.path.join(shared_folder_shortcut, item)
        dest_item = os.path.join(my_workspace_folder, item)

        # Copy directories (sub-folders)
        if os.path.isdir(source_item):
            if not os.path.exists(dest_item):
                shutil.copytree(source_item, dest_item)
                print(f"📁 Copied folder: {item}")
            else:
                print(f"⏭️ Skipped folder (already exists): {item}")

        # Copy standard files
        else:
            # We check if it exists so we don't accidentally overwrite your work later
            if not os.path.exists(dest_item):
                shutil.copy2(source_item, dest_item)
                print(f"📄 Copied file: {item}")
            else:
                print(f"⏭️ Skipped file (already exists): {item}")

    print(f"\n✅ Success! You can now work with your files in the '{my_workspace_folder}' folder.")
else:
    print(f"❌ ERROR: Could not find '{shared_folder_shortcut}'.")
    print("Please make sure you have added a shortcut to 'Mtdk 2026' directly into your 'My Drive'.")
```
